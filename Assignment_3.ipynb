2. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.The random variable $Y$ maps elements to $\{0,1\}$. We look at the pre-images of its possible values:$Y^{-1}(\{0\}) = \{\omega_1, \omega_2\}$$Y^{-1}(\{1\}) = \{\omega_3, \omega_4\}$The $\sigma$-algebra generated by $Y$ consists of the empty set, the full space, and these pre-images along with their complements:$$\sigma(Y) = \Big\{ \emptyset, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\}, \Omega \Big\}$$

3. Explain why $E[X \mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets $\{\omega_1, \omega_2\}$ and $\{\omega_3, \omega_4\}$.By definition, the conditional expectation $E[X \mid Y]$ is a random variable that represents our best estimate of $X$ given the information contained in $Y$. Because it is entirely determined by the value that $Y$ takes, it must be measurable with respect to the information context generated by $Y$, which is $\sigma(Y)$.Any function that is $\sigma(Y)$-measurable must be constant over the "atoms" (the smallest non-empty building blocks) of $\sigma(Y)$. Since the atoms of $\sigma(Y)$ are exactly $\{\omega_1, \omega_2\}$ and $\{\omega_3, \omega_4\}$, $E[X \mid Y]$ cannot distinguish between elements within the same atom and must yield a uniform, constant value across each set.

Compute $E[X \mid Y]$ explicitly and show that $E[X \mid Y] = \frac{1}{6}\mathbb{1}_{\{Y=0\}} + \frac{1}{2}\mathbb{1}_{\{Y=1\}}$.We evaluate the conditional expectation on each atom of $\sigma(Y)$:Case 1: On the set $\{Y=0\} = \{\omega_1, \omega_2\}$$$E[X \mid Y=0] = \frac{E[X \cdot \mathbb{1}_{\{\omega_1, \omega_2\}}]}{P(\{\omega_1, \omega_2\})} = \frac{X(\omega_1)P(\{\omega_1\}) + X(\omega_2)P(\{\omega_2\})}{P(\{\omega_1\}) + P(\{\omega_2\})}$$$$E[X \mid Y=0] = \frac{0 \cdot 0.50 + 1 \cdot 0.10}{0.50 + 0.10} = \frac{0.10}{0.60} = \frac{1}{6}$$Case 2: On the set $\{Y=1\} = \{\omega_3, \omega_4\}$$$E[X \mid Y=1] = \frac{E[X \cdot \mathbb{1}_{\{\omega_3, \omega_4\}}]}{P(\{\omega_3, \omega_4\})} = \frac{X(\omega_3)P(\{\omega_3\}) + X(\omega_4)P(\{\omega_4\})}{P(\{\omega_3\}) + P(\{\omega_4\})}$$$$E[X \mid Y=1] = \frac{0 \cdot 0.20 + 1 \cdot 0.20}{0.20 + 0.20} = \frac{0.20}{0.40} = \frac{1}{2}$$Combining these piecewise results using indicator functions yields:$$E[X \mid Y] = \frac{1}{6}\mathbb{1}_{\{Y=0\}} + \frac{1}{2}\mathbb{1}_{\{Y=1\}}$$

5. Verify the defining property of conditional expectation by checking that for every set $A \in \sigma(Y)$, $\int_A E[X \mid Y] \, dP = \int_A X \, dP$.
We test this condition for all four elements in $\sigma(Y)$:

(1) For $A = \emptyset$: Both sides are trivially $0$.

(2) For $A = \Omega$:$$\int_{\Omega} E[X \mid Y] \, dP = \frac{1}{6}P(\{\omega_1, \omega_2\}) + \frac{1}{2}P(\{\omega_3, \omega_4\}) = \frac{1}{6}(0.60) + \frac{1}{2}(0.40) = 0.10 + 0.20 = 0.30$$$$\int_{\Omega} X \, dP = 0 \cdot P(\{\omega_1\}) + 1 \cdot P(\{\omega_2\}) + 0 \cdot P(\{\omega_3\}) + 1 \cdot P(\{\omega_4\}) = 0.10 + 0.20 = 0.30$$Both sides equal $0.30$.

(3) For $A = \{\omega_1, \omega_2\}$:$$\int_A E[X \mid Y] \, dP = \frac{1}{6} P(\{\omega_1, \omega_2\}) = \frac{1}{6}(0.60) = 0.10$$$$\int_A X \, dP = X(\omega_1)P(\{\omega_1\}) + X(\omega_2)P(\{\omega_2\}) = 0 + 1 \cdot 0.10 = 0.10$$Both sides equal $0.10$.

(4) For $A = \{\omega_3, \omega_4\}$:$$\int_A E[X \mid Y] \, dP = \frac{1}{2} P(\{\omega_3, \omega_4\}) = \frac{1}{2}(0.40) = 0.20$$$$\int_A X \, dP = X(\omega_3)P(\{\omega_3\}) + X(\omega_4)P(\{\omega_4\}) = 0 + 1 \cdot 0.20 = 0.20$$Both sides equal $0.20$.The defining property holds true across all cases.

8. Calculate the Mean Square Error (MSE).
When predicting $X$ using an explicit observation $Y$, the optimal estimator minimizing mean square error is the conditional expectation $\hat{X} = E[X \mid Y]$.
$$\text{MSE} = E\left[(X - E[X \mid Y])^2\right] = \sum_{\omega_i \in \Omega} (X(\omega_i) - E[X \mid Y](\omega_i))^2 P(\{\omega_i\})
$$For $\omega_1$: $(0 - \frac{1}{6})^2 (0.50) = \frac{1}{36} \cdot \frac{1}{2} = \frac{1}{72}$

For $\omega_2$: $(1 - \frac{1}{6})^2 (0.10) = \frac{25}{36} \cdot \frac{1}{10} = \frac{25}{360} = \frac{5}{72}$

For $\omega_3$: $(0 - \frac{1}{2})^2 (0.20) = \frac{1}{4} \cdot \frac{1}{5} = \frac{1}{20}$

For $\omega_4$: $(1 - \frac{1}{2})^2 (0.20) = \frac{1}{4} \cdot \frac{1}{5} = \frac{1}{20}$
$$\text{MSE} = \left(\frac{1}{72} + \frac{5}{72}\right) + \left(\frac{1}{20} + \frac{1}{20}\right) = \frac{6}{72} + \frac{2}{20} = \frac{1}{12} + \frac{1}{10} = \frac{5 + 6}{60} = \mathbf{\frac{11}{60} \approx 0.1833}$$

9. How would the Mean Square Error (MSE) change if the bank decided to ignore the customer risk class $Y$ and instead used the global average fraud rate ($E[X] = 0.30$)? Calculate this error and explain its relationship to the Law of Total Variance.
If we ignore $Y$, our constant estimate becomes $\hat{X}_{const} = E[X] = 0.30$. The new error is simply the total variance of $X$:
$$\text{MSE}_{\text{constant}} = \text{Var}(X) = E[X^2] - (E[X])^2
$$Since $X \in \{0,1\}$, $X^2 = X$, so $E[X^2] = E[X] = 0.30$.
$$\text{Var}(X) = 0.30 - (0.30)^2 = 0.30 - 0.09 = \mathbf{0.21}
$$Relationship to the Law of Total Variance:

The Law of Total Variance states:

$\text{Var}(X) = E[\text{Var}(X \mid Y)] + \text{Var}(E[X \mid Y])$

$$E[\text{Var}(X \mid Y)]  is the remaining internal variance after observing Y$, which corresponds exactly to the minimized $\text{MSE}$ from Question 8: $\frac{11}{60} \approx 0.1833$.
$\text{Var}(E[X \mid Y])$ is the variance explained by knowing $Y$:
$$E\left[(E[X \mid Y])^2\right] = \left(\frac{1}{6}\right)^2(0.60) + \left(\frac{1}{2}\right)^2(0.40) = \frac{1}{36}\left(\frac{3}{5}\right) + \frac{1}{4}\left(\frac{2}{5}\right) = \frac{1}{60} + \frac{6}{60} = \frac{7}{60}$$
$$\text{Var}(E[X \mid Y]) = \frac{7}{60} - (0.30)^2 = \frac{7}{60} - \frac{9}{100} = \frac{35 - 27}{300} = \frac{8}{300} = \frac{2}{75} \approx 0.0267
$$Checking the identity breakdown:
$0.21 = 0.1833 + 0.0267
$Ignoring $Y$ increases the prediction error because you lose the variance explained by the stratification model ($\text{Var}(E[X \mid Y]) = 0.0267$), raising your structural error penalty from $0.1833$ to $0.21$.

10. Information Theory Calculations
(a) Compute the entropy $H(X)$The marginal distribution of $X$ is $P(X=1) = 0.10 + 0.20 = 0.30$, and $P(X=0) = 0.70$.$$H(X) = -0.30 \log_2(0.30) - 0.70 \log_2(0.70)$$$$H(X) \approx -0.30(-1.737) - 0.70(-0.515) = 0.521 + 0.361 = \mathbf{0.881 \text{ bits}}$$

(b) Compute the conditional entropy $H(X \mid Y)$Using the conditional probabilities from Question 6:$$H(X \mid Y) = P(Y=0)H(X \mid Y=0) + P(Y=1)H(X \mid Y=1)$$$H(X \mid Y=0) = -\frac{5}{6}\log_2\left(\frac{5}{6}\right) - \frac{1}{6}\log_2\left(\frac{1}{6}\right) \approx 0.650 \text{ bits}$$H(X \mid Y=1) = -\frac{1}{2}\log_2\left(\frac{1}{2}\right) - \frac{1}{2}\log_2\left(\frac{1}{2}\right) = 1.000 \text{ bit}$$$H(X \mid Y) = 0.60(0.650) + 0.40(1.000) = 0.390 + 0.400 = \mathbf{0.790 \text{ bits}}$$

(c) Calculate the Mutual Information $I(X; Y)$$$I(X; Y) = H(X) - H(X \mid Y) = 0.881 - 0.790 = \mathbf{0.091 \text{ bits}}$$

(d) What percentage of the total uncertainty about fraud is removed by knowing the customer's risk class?$$\text{Percentage Uncertainty Removed} = \frac{I(X; Y)}{H(X)} \times 100\% = \frac{0.091}{0.881} \times 100\% \approx \mathbf{10.33\%}$$
