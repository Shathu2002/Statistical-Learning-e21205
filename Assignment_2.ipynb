# Q1) The Probability of Shooting at a Target

Let the disk have radius $R=10$. Event $A$ occurs if the point lies within $1$ unit of the boundary, meaning its distance from the center $r$ satisfies $9 \le r \le 10$.

### 1. Uniform Distribution with Respect to Area
In this model, the probability is the ratio of the areas:
* $Area_{total} = \pi(10)^2 = 100\pi$
* $Area_{A} = \pi(10^2) - \pi(9^2) = 100\pi - 81\pi = 19\pi$
$$P(A) = \frac{19\pi}{100\pi} = \mathbf{0.19}$$

### 2. Uniform Distribution of Distance and Direction
In this model, $r$ is chosen uniformly from the interval $[0, 10]$:
$$P(A) = P(9 \le r \le 10) = \frac{10 - 9}{10 - 0} = \mathbf{0.1}$$

### 3. Why the Answers Differ
The answers differ because "uniformly distributed" depends on the choice of variables. The area model accounts for the fact that outer rings have more surface area than inner rings, while the distance model treats all radial intervals as equally likely, regardless of their physical size.

# Q2) Bertrand's Paradox

Bertrand's Paradox demonstrates that "at random" is ambiguous for infinite sample spaces. For an inscribed equilateral triangle with side length $L = R\sqrt{3}$, we find:

* **Random Endpoints:** Fix one point; the chord exceeds $L$ if the second point is in the middle $1/3$ arc. $P = \mathbf{1/3}$.
* **Random Radius:** The chord exceeds $L$ if its midpoint is closer to the center than $R/2$. $P = \mathbf{1/2}$.
* **Random Midpoint:** The chord exceeds $L$ if the midpoint falls within a concentric circle of radius $R/2$. $P = \frac{\pi(R/2)^2}{\pi R^2} = \mathbf{1/4}$.
