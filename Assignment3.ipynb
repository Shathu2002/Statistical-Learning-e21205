 Why $X$ is not $\sigma(Y)$-measurable

 A random variable $X$ is $\sigma(Y)$-measurable if and only if its value can be completely and uniquely determined solely by observing the value of $Y$.
Mathematically, this requires a functional mapping $g$ such that $X = g(Y)$.

Mutual InformationCalculate $I(X; Y)$ and find the percentage of uncertainty removedThe Mutual Information measures the shared information between the variables:$$I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2}\ln(2\pi e) - \frac{1}{2}\ln\left(2\pi e \cdot \frac{4}{5}\right)$$$$I(X; Y) = \frac{1}{2}\ln\left( \frac{2\pi e}{2\pi e \cdot 0.8} \right) = \frac{1}{2}\ln\left(\frac{5}{4}\right) = \frac{1}{2}\ln(1.25) \approx 0.1116 \text{ nats}$$The percentage of initial temperature uncertainty removed by observing the power meter is:$$\text{Percentage Uncertainty Removed} = \frac{I(X;Y)}{h(X)} \times 100\% = \frac{\frac{1}{2}\ln(1.25)}{\frac{1}{2}\ln(2\pi e)} \times 100\% = \frac{0.2231}{2.8379} \times 100\% \approx 7.86\%$$How would this change if mechanical noise $Q$ had a variance of $100$ instead of $4$?If $\text{Var}(Q) = 100$, the updated total variance becomes $\text{Var}(Y) = \text{Var}(X) + \text{Var}(Q) = 1 + 100 = 101$.The new conditional variance of $X$ given $Y$ is:$$\sigma_{X|Y,\text{new}}^2 = 1 - \frac{1^2}{101} = \frac{100}{101}$$The new Mutual Information becomes:$$I(X; Y)_{\text{new}} = \frac{1}{2}\ln\left( \frac{\sigma_X^2}{\sigma_{X|Y,\text{new}}^2} \right) = \frac{1}{2}\ln\left( \frac{1}{\frac{100}{101}} \right) = \frac{1}{2}\ln\left(\frac{101}{100}\right) = \frac{1}{2}\ln(1.01) \approx 0.00497 \text{ nats}$$The new percentage of uncertainty removed drops to:$$\text{Percentage Uncertainty Removed}_{\text{new}} = \frac{\frac{1}{2}\ln(1.01)}{\frac{1}{2}\ln(2\pi e)} \times 100\% = \frac{0.00995}{2.8379} \times 100\% \approx 0.35\%$$As noise increases significantly, the observation provides almost zero information about the true temperature.

7. Diagnostic Interpretation

Why $Y$ is a biased estimator of $X$

An estimator $\hat{X}$ is unbiased if $E[\hat{X}] = E[X]$ under all localized conditions. If we use the manager's proposal ($\hat{X} = Y$), then given a specific real state where the true temperature deviation is $X = x$, the expected value of our estimate is:$$E[Y \mid X = x] = E[X + Q \mid X = x] = x + E[Q] = x + 0 = x$$While it appears unbiased globally, the true task is predicting $X$ given an observation $Y$. If we evaluate the expectation of the manager's estimator under our observed reality, the conditional bias is exposed:$$E[Y - X \mid Y] = Y - E[X \mid Y] = Y - \frac{1}{5}Y = \frac{4}{5}Y \neq 0

$$Because this value does not equal zero for non-zero observations, $Y$ functions as a fundamentally flawed system predictor that overestimates the scale of structural deviations.

Why $E[X \mid Y]$ "shrinks" the measurement toward zero

The optimal Minimum Mean Square Error (MMSE) estimator is $E[X \mid Y] = \frac{1}{5}Y = 0.2Y$.
This estimator intentionally "shrinks" the raw reading $Y$ toward the prior mean ($\mu_X = 0$) because it recognizes that any extreme reading in $Y$ is statistically more likely to be caused by a random spike in the high-variance noise component ($Q \sim \mathcal{N}(0,4)$) than by a spike in the low-variance signal component ($X \sim \mathcal{N}(0,1)$).
