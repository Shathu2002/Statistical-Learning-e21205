# Q1) The Probability of Shooting at a Target

### 1. Uniform Distribution with Respect to Area
In this model, the probability is based on the ratio of the area of the event to the total area of the disk (Radius $R = 10$).

* **Total Area:** $A_{total} = \pi \times 10^2 = 100\pi$
* **Event $A$:** The point lies within 1 unit of the boundary, meaning the radius $r$ is between 9 and 10.
* **Area of $A$:** This is the area of the outer ring (annulus):
    $$Area(A) = \pi(10)^2 - \pi(9)^2 = 100\pi - 81\pi = 19\pi$$
* **Probability:**
    $$P(A) = \frac{19\pi}{100\pi} = 0.19$$

---

### 2. Uniform Distribution of Distance and Direction
Here, the distance $r$ is chosen uniformly from $[0, 10]$. The direction $\theta$ is independent and does not affect the radial distance.

* **Sample Space:** The interval $[0, 10]$ for the variable $r$.
* **Event $A$:** The distance $r$ satisfies $9 \le r \le 10$.
* **Probability:** Since $r$ is uniform:
    $$P(A) = \frac{\text{Length of favorable interval}}{\text{Total length}} = \frac{10 - 9}{10 - 0} = \frac{1}{10} = 0.1$$

---

### 3. Explanation of the Difference
The answers differ because the **probability density** is distributed differently:
* In the **Area Model**, a "random point" is equally likely to be anywhere on the surface. Because there is more physical area near the edge than near the center, $P(A)$ is higher.
* In the **Distance Model**, a "random point" is defined by its distance from the center. This model ignores the fact that rings further out have more area, effectively "compressing" the points toward the edge less than the area model does.

# Q2) Bertrand's Paradox

The paradox arises because "choosing a chord at random" is not a well-defined operation. Depending on the geometric parameter chosen to be uniform, we get different results. Let $L$ be the side of the inscribed equilateral triangle ($L = R\sqrt{3}$).



| Model | Method of Selection | Logic | Probability |
| :--- | :--- | :--- | :--- |
| **1. Random Endpoints** | Choose two points on the circumference. | Rotational symmetry allows us to fix one point. The chord is longer than $L$ if the second point falls on the middle $1/3$ of the circle. | $P = \frac{1}{3}$ |
| **2. Random Radius** | Choose a radius and a point on it. | A chord is longer than $L$ if its midpoint is closer to the center than $R/2$. On a radius of length $R$, this is the inner half. | $P = \frac{1}{2}$ |
| **3. Random Midpoint** | Choose a point anywhere in the disk. | A chord is longer than $L$ if the chosen midpoint falls inside a concentric circle with radius $R/2$. | $P = \frac{\text{Area}_{inner}}{\text{Area}_{total}} = \frac{\pi(R/2)^2}{\pi R^2} = \frac{1}{4}$ |

### Conclusion
Bertrand's Paradox demonstrates that without a strictly defined **transformation group** (like saying the distribution must be invariant under translations and rotations), the term "at random" is ambiguous in infinite sample spaces.

# Q3) Transformer Failures

### 1. Construct a Probability Space $(\Omega, \mathcal{F}, P)$
To model the number of failures, we define the components of the probability space:
* **Sample Space ($\Omega$):** The set of all possible outcomes. Since the number of failures is between 0 and 3, $\Omega = \{0, 1, 2, 3\}$.
* **Sigma-algebra ($\mathcal{F}$):** The set of all possible events. For a finite sample space, we use the power set $\mathcal{F} = \mathcal{P}(\Omega)$, which contains all $2^4 = 16$ possible subsets of $\Omega$.
* **Probability Measure ($P$):** Defined by the given values:
  $P(\{0\}) = 0.50, P(\{1\}) = 0.30, P(\{2\}) = 0.15, P(\{3\}) = 0.05$.

### 2. Define $X$ as a Random Variable
The random variable $X$ is a function $X: \Omega \to \mathbb{R}$ that maps each outcome to its numerical value. In this case, it is the identity mapping:
$$X(\omega) = \omega \quad \text{for all } \omega \in \Omega$$

### 3. Probability Mass Function (PMF)
The PMF $p_X(x) = P(X=x)$ is given by:
| $x$ | $p_X(x)$ |
| :--- | :--- |
| 0 | 0.50 |
| 1 | 0.30 |
| 2 | 0.15 |
| 3 | 0.05 |

---

### 4. Compute the Expectation $E[X]$
The expectation is the weighted average of the outcomes:
$$E[X] = \sum_{x=0}^{3} x \cdot P(X=x)$$
$$E[X] = (0 \cdot 0.50) + (1 \cdot 0.30) + (2 \cdot 0.15) + (3 \cdot 0.05)$$
$$E[X] = 0 + 0.30 + 0.30 + 0.15 = \mathbf{0.75}$$

---

### 5. Compute the Variance $Var(X)$
First, we find $E[X^2]$:
$$E[X^2] = \sum x^2 \cdot P(X=x) = (0^2 \cdot 0.50) + (1^2 \cdot 0.30) + (2^2 \cdot 0.15) + (3^2 \cdot 0.05)$$
$$E[X^2] = 0 + 0.30 + 0.60 + 0.45 = 1.35$$
Now, use the formula $Var(X) = E[X^2] - (E[X])^2$:
$$Var(X) = 1.35 - (0.75)^2 = 1.35 - 0.5625 = \mathbf{0.7875}$$

---

### 6. Find $P(X \ge 1)$ and $P(X \ge 2)$
* **$P(X \ge 1)$:** Using the complement rule:
  $$P(X \ge 1) = 1 - P(X = 0) = 1 - 0.50 = \mathbf{0.50}$$
* **$P(X \ge 2)$:** Summing the probabilities:
  $$P(X \ge 2) = P(X=2) + P(X=3) = 0.15 + 0.05 = \mathbf{0.20}$$

---

### 7. Practical Meaning
* **Expectation ($0.75$):** This represents the **long-term average** number of transformer failures per day. If the district monitors failures over many years, the average number of failures per day will converge to 0.75.
* **Variance ($0.7875$):** This measures the **volatility or spread** of the failure reports. A higher variance would indicate that the number of failures fluctuates significantly from day to day, making it harder for emergency crews to predict staffing needs.

# Q4) Probability of Defective Items

### 1. Construct a Probability Space $(\Omega, \mathfrak{F}, P)$
* **Sample Space ($\Omega$):** Since $X$ represents a proportion, $\Omega = [0, 1]$.
* **Sigma-algebra ($\mathfrak{F}$):** The Borel $\sigma$-algebra on $[0, 1]$, denoted as $\mathcal{B}([0, 1])$.
* **Probability Measure ($P$):** For any event $A \in \mathfrak{F}$, $P(A) = \int_A f(x) \, dx$, where $f(x)$ is the Beta density function.

### 2. Define $X$ as a Continuous Random Variable
$X$ is the identity function $X: \Omega \to [0, 1]$ such that $X(\omega) = \omega$. It is continuous because its cumulative distribution function (CDF) is continuous.

### 3. Probability Density Function (PDF)
The PDF of $X \sim \text{Beta}(\alpha=2, \beta=18)$ is:
$$f(x) = \begin{cases} \frac{1}{B(2, 18)} x^1 (1-x)^{17}, & 0 < x < 1 \\ 0, & \text{otherwise} \end{cases}$$

### 4. Verify that $f$ is a Valid PDF
A valid PDF must satisfy $\int_{-\infty}^{\infty} f(x) \, dx = 1$. Using the definition of the Beta function $B(\alpha, \beta) = \int_0^1 x^{\alpha-1}(1-x)^{\beta-1} dx$:
$$\int_0^1 \frac{1}{B(2, 18)} x^{2-1}(1-x)^{18-1} dx = \frac{1}{B(2, 18)} \int_0^1 x^1(1-x)^{17} dx = \frac{B(2, 18)}{B(2, 18)} = 1$$

### 5. Compute the Expectation $E[X]$
For a Beta distribution, $E[X] = \frac{\alpha}{\alpha + \beta}$:
$$E[X] = \frac{2}{2 + 18} = \frac{2}{20} = \mathbf{0.10}$$

### 6. Compute the Variance $Var(X)$
For a Beta distribution, $Var(X) = \frac{\alpha\beta}{(\alpha + \beta)^2(\alpha + \beta + 1)}$:
$$Var(X) = \frac{2 \cdot 18}{(20)^2(21)} = \frac{36}{400 \cdot 21} = \frac{36}{8400} = \frac{3}{700} \approx \mathbf{0.00428}$$

---

### 7. Probability $P(X \le 0.10)$
Leaving the answer in Beta integral form:
$$P(X \le 0.10) = \frac{1}{B(2, 18)} \int_0^{0.10} x(1-x)^{17} dx$$

### 8. Probability $P(0.05 \le X \le 0.15)$
Leaving the answer in integral form:
$$P(0.05 \le X \le 0.15) = \frac{1}{B(2, 18)} \int_{0.05}^{0.15} x(1-x)^{17} dx$$

---

### 9. Why a Beta Distribution is Reasonable
The Beta distribution is ideal for modeling proportions because:
* Its **support is bounded** between 0 and 1, matching the physical constraints of a percentage or proportion.
* It is **flexible**; by adjusting $\alpha$ and $\beta$, we can model various shapes of uncertainty (skewed, symmetric, or flat).

### 10. Interpretation in Production Quality
* **Expectation ($0.10$):** This is the **average defective rate**. On a typical day, the manufacturer expects about 10% of the items to be defective.
* **Variance ($\approx 0.0043$):** This represents the **consistency of the process**. A low variance suggests that the defective rate stays very close to 10% day-to-day, indicating a stable (though perhaps not perfect) manufacturing process.

# Q5) Expected Project Completion Time

Given parameters for the Triangular distribution: $a = 4$ (min), $b = 10$ (max), and $c = 6$ (mode).

### 1. Specify the Support of $X$
The support of the random variable $X$ is the interval of all possible completion times:
$$\text{Support} = [a, b] = [4, 10]$$

### 2. Probability Density Function $f_X(x)$
The PDF for a triangular distribution is defined piecewise:
$$f_X(x) = \begin{cases}
\frac{2(x-a)}{(b-a)(c-a)} & \text{for } a \le x < c \\
\frac{2(b-x)}{(b-a)(b-c)} & \text{for } c \le x \le b \\
0 & \text{otherwise}
\end{cases}$$
Substituting the values $a=4, b=10, c=6$:
$$f_X(x) = \begin{cases}
\frac{2(x-4)}{(6)(2)} = \frac{x-4}{6} & \text{for } 4 \le x < 6 \\
\frac{2(10-x)}{(6)(4)} = \frac{10-x}{12} & \text{for } 6 \le x \le 10
\end{cases}$$

### 3. Verify that $f_X$ Integrates to 1
The integral over the support is the area of the triangle with base $(b-a)$ and height $h = \frac{2}{b-a}$:
$$\int_{4}^{10} f_X(x)dx = \text{Area} = \frac{1}{2} \times \text{base} \times \text{height} = \frac{1}{2} \times (10-4) \times \frac{2}{10-4} = 1$$

### 4. Compute the Expectation $E[X]$
For a Triangular distribution, $E[X] = \frac{a+b+c}{3}$:
$$E[X] = \frac{4 + 10 + 6}{3} = \frac{20}{3} \approx \mathbf{6.67} \text{ hours}$$

### 5. Compute the Variance $Var(X)$
The formula for variance is $Var(X) = \frac{a^2 + b^2 + c^2 - ab - ac - bc}{18}$:
$$Var(X) = \frac{4^2 + 10^2 + 6^2 - (4 \cdot 10) - (4 \cdot 6) - (10 \cdot 6)}{18} = \frac{16 + 100 + 36 - 40 - 24 - 60}{18} = \frac{28}{18} = \mathbf{1.55}$$

---

### 6. Interpret $a, b, \text{ and } c$
* **$a=4$:** The absolute best-case scenario (optimistic time).
* **$b=10$:** The absolute worst-case scenario (pessimistic time).
* **$c=6$:** The most frequent or "likely" time required based on experience.

### 7. Why a Triangular Distribution is Reasonable
It is a "subjective" distribution used when limited data exists. It only requires three intuitive estimates (min, max, mode), making it easier for experts to provide inputs than complex distributions like Normal or Gamma.

---

### 8. Probability $P(X \le 5)$
Since $5 < c$, we use the first part of the PDF:
$$P(X \le 5) = \int_{4}^{5} \frac{x-4}{6} dx = \left[ \frac{(x-4)^2}{12} \right]_4^5 = \frac{1^2}{12} - 0 = \mathbf{0.0833}$$

### 9. Probability $P(5 \le X \le 8)$
$$P(5 \le X \le 8) = P(X \le 8) - P(X \le 5)$$
$P(X \le 8)$ involves both parts of the PDF (or use $1 - P(X > 8)$):
$$P(X > 8) = \int_{8}^{10} \frac{10-x}{12} dx = \left[ -\frac{(10-x)^2}{24} \right]_8^{10} = 0 - (-\frac{4}{24}) = \frac{1}{6} \approx 0.1667$$
$$P(5 \le X \le 8) = 1 - P(X > 8) - P(X < 5) = 1 - 0.1667 - 0.0833 = \mathbf{0.75}$$

### 10. Project Planning and Sensitivity Analysis
This model helps planners determine **contingency buffers**. If the probability of exceeding 9 hours is high, they might allocate more resources. In **sensitivity analysis**, planners can change $b$ (the pessimistic time) to see how much the average completion time ($E[X]$) shifts, identifying which tasks are most critical to the total project timeline.

# Q6) Expected Project Completion Time Using PERT

Given parameters: $a = 8$ (optimistic), $b = 20$ (pessimistic), and $c = 11$ (most likely).

### 1. Compute Shape Parameters $\alpha$ and $\beta$
Using the provided formulas:
* $$\alpha = 1 + 4\left(\frac{c - a}{b - a}\right) = 1 + 4\left(\frac{11 - 8}{20 - 8}\right) = 1 + 4\left(\frac{3}{12}\right) = 1 + 1 = \mathbf{2}$$
* $$\beta = 1 + 4\left(\frac{b - c}{b - a}\right) = 1 + 4\left(\frac{20 - 11}{20 - 8}\right) = 1 + 4\left(\frac{9}{12}\right) = 1 + 3 = \mathbf{4}$$

### 2. Specify the Support of $X$
The support of the PERT distribution is the interval between the shortest and longest possible times:
$$\text{Support} = [a, b] = [8, 20]$$

### 3. Probability Density Function $f_X(x)$
Substituting $\alpha=2, \beta=4, a=8, b=20$ into the density function:
$$f_X(x) = \frac{(x - 8)^{2-1}(20 - x)^{4-1}}{B(2, 4)(20 - 8)^{2+4-1}} = \frac{(x - 8)(20 - x)^3}{B(2, 4) \cdot 12^5}, \quad 8 \le x \le 20$$

### 4. Probability Space $(\Omega, \mathfrak{F}, P)$
* **Sample Space ($\Omega$):** $[8, 20]$
* **Sigma-algebra ($\mathfrak{F}$):** Borel $\sigma$-algebra $\mathcal{B}([8, 20])$
* **Probability Measure ($P$):** $P(A) = \int_A f_X(x) \, dx$ for $A \in \mathfrak{F}$.

### 5. Define $X$ as a Random Variable
$X$ is the identity map $X(\omega) = \omega$ for $\omega \in \Omega$, representing the number of days to complete the task.

---

### 6. Show that $E[X] = \frac{a + 4c + b}{6}$
For a Beta distribution on $[0,1]$ with parameters $(\alpha, \beta)$, the mean is $\frac{\alpha}{\alpha + \beta}$. For a PERT distribution scaled to $[a, b]$, the mean is $a + (b-a)\frac{\alpha}{\alpha+\beta}$.
Substituting $\alpha$ and $\beta$ formulas:
$$E[X] = a + (b-a)\frac{1 + 4\frac{c-a}{b-a}}{(1 + 4\frac{c-a}{b-a}) + (1 + 4\frac{b-c}{b-a})} = a + (b-a)\frac{1 + 4\frac{c-a}{b-a}}{2 + 4(\frac{c-a+b-c}{b-a})} = a + (b-a)\frac{1 + 4\frac{c-a}{b-a}}{6}$$
$$E[X] = \frac{6a + (b-a) + 4(c-a)}{6} = \frac{6a + b - a + 4c - 4a}{6} = \frac{\mathbf{a + 4c + b}}{\mathbf{6}}$$
Calculation: $E[X] = \frac{8 + 4(11) + 20}{6} = \frac{72}{6} = \mathbf{12}$

### 7. Compute Variance $Var(X)$
Using the approximation formula:
$$Var(X) = \frac{(12 - 8)(20 - 12)}{7} = \frac{4 \cdot 8}{7} = \frac{32}{7} \approx \mathbf{4.57}$$

### 8. Compute Standard Deviation $\sigma$
$$\sigma = \sqrt{Var(X)} = \sqrt{\frac{32}{7}} \approx \mathbf{2.14}$$

---

### 9. PERT vs. Triangular Distribution
The **PERT distribution** is often more appropriate than the **Triangular distribution** because it is "smoother" and places less weight on the extreme tails (the optimistic and pessimistic values). The Triangular distribution mean is $(a+b+c)/3$. In this case, $13$ vs. PERT's $12$. PERT more effectively emphasizes the "most likely" value $c$ via the weight of 4 in the numerator.

### 10. Practical Meaning
* **$a, b, c$:** Define the boundaries and peak of professional judgment/risk.
* **$E[X] = 12$:** The average time the project manager should budget for this task over many similar instances.
* **$\sigma \approx 2.14$:** Represents the uncertainty. A higher $\sigma$ would indicate a "riskier" task where the actual completion date could vary widely from the 12-day estimate.

# Q7) If the PERT Expectation is to Match Our Intuition

Given: $a = 5$, $c = 11$. The condition is that the expected time $E[X]$ must be equal to the most likely time $c$.

### 1. Determine the value of the maximum time $b$
The formula for the expectation of a PERT distribution is:
$$E[X] = \frac{a + 4c + b}{6}$$

According to the problem, $E[X] = c$. We set up the equation:
$$c = \frac{a + 4c + b}{6}$$

Multiply both sides by 6:
$$6c = a + 4c + b$$

Isolate $b$:
$$b = 6c - 4c - a$$
$$b = 2c - a$$

Substitute the given values $a = 5$ and $c = 11$:
$$b = 2(11) - 5$$
$$b = 22 - 5 = \mathbf{17}$$

**Result:** For the mean to equal the mode, the distribution must be symmetric, which occurs when $b = 17$.

### 2. What happens to the variance?
When $E[X] = c$, the distribution becomes perfectly symmetric.
* Using the approximation $Var(X) = \frac{(E[X] - a)(b - E[X])}{7}$:
* Substitute $E[X] = 11$:
$$Var(X) = \frac{(11 - 5)(17 - 11)}{7} = \frac{6 \cdot 6}{7} = \frac{36}{7} \approx \mathbf{5.14}$$

If the distribution were skewed (e.g., if $b$ were much larger), the variance would typically increase as the "pessimistic" tail pulls the spread wider. In this specific case, setting $E[X]=c$ minimizes the variance relative to the range by centering the bulk of the probability mass.

# Q8) Probability of Being a Girl

This is a classic problem in conditional probability. Let $G$ represent a girl and $B$ represent a boy.

### 1. Define the Sample Space
For a family with two children, the possible ordered outcomes are:
$$\Omega = \{BB, BG, GB, GG\}$$
Since each child is equally likely to be a boy or girl, each outcome has a probability of $1/4$.

### 2. Define the Events
* **Event $A$:** Both children are girls. $A = \{GG\}$.
* **Event $B$:** At least one child is a girl. $B = \{BG, GB, GG\}$.

### 3. Apply Conditional Probability
We want to find $P(A | B)$, the probability that both are girls given at least one is a girl.
$$P(A | B) = \frac{P(A \cap B)}{P(B)}$$

* $A \cap B = \{GG\}$ (The intersection of "both girls" and "at least one girl" is just "both girls").
* $P(A \cap B) = \frac{1}{4}$
* $P(B) = \frac{3}{4}$

$$P(A | B) = \frac{1/4}{3/4} = \mathbf{\frac{1}{3}}$$

**Conclusion:** The probability is $1/3$ (approximately $33.3\%$). This often surprises people because they assume the gender of the "other" child is $1/2$, but the information "at least one" eliminates only the $BB$ outcome, leaving three equally likely possibilities.

```python
import matplotlib.pyplot as plt

# Data for X
x_labels = ['Good (0)', 'Defective (1)']
x_probs = [0.

# Q10) Information Available in Continuous Random Variables

Given a chemical plant monitoring system with temperature deviation $T \sim N(0, 1)$ and pressure deviation $P \sim N(0, 4)$, where $T$ and $P$ are independent.

### 1. The Probability Space $(\Omega, \mathfrak{F}, P)$
* **Sample Space:** $\Omega = \mathbb{R}^2$.
* **Sigma-algebra:** $\mathfrak{F} = \mathcal{B}(\mathbb{R}^2)$ (the Borel $\sigma$-algebra on $\mathbb{R}^2$).
* **Probability Measure:** Induced by the joint density function:
  $$f_{T,P}(t,p) = \frac{1}{4\pi} \exp\left( -\frac{t^2}{2} - \frac{p^2}{8} \right)$$

### 2. $\sigma$-algebra generated by $X$, $\sigma(X)$
$X$ only measures temperature. Therefore, $\sigma(X)$ consists of all events of the form $B \times \mathbb{R}$, where $B \in \mathcal{B}(\mathbb{R})$. It is a "strip-based" $\sigma$-algebra that cannot distinguish between different pressure values.

### 3. $\sigma$-algebra generated by $Y$, $\sigma(Y)$
$Y$ measures both temperature and pressure. Since $Y$ is the identity map on $\mathbb{R}^2$, it generates the full Borel $\sigma$-algebra:
$$\sigma(Y) = \mathcal{B}(\mathbb{R}^2)$$

### 4. Show that $\sigma(X) \subseteq \sigma(Y)$
Every set in $\sigma(X)$ (a cylinder set or "strip") is by definition a Borel set in $\mathbb{R}^2$. However, $\sigma(Y)$ contains many sets (like small circles or rectangles) that cannot be expressed as simple strips along the $T$-axis. Thus, $Y$ contains more information because it distinguishes outcomes in two dimensions, whereas $X$ collapses everything into one.

---

### 5. Marginal Density of $X$
Since $X(t,p) = t$, its density is simply the density of $T$:
$$f_X(x) = \frac{1}{\sqrt{2\pi}} \exp\left( -\frac{x^2}{2} \right)$$

### 6. Marginal Density of $Y$
$Y$ is the vector $(T, P)$. Since they are independent, the joint density is the product of their marginals:
$$f_Y(t,p) = f_T(t) \cdot f_P(p) = \left[ \frac{1}{\sqrt{2\pi}} e^{-t^2/2} \right] \cdot \left[ \frac{1}{2\sqrt{2\pi}} e^{-p^2/8} \right] = \frac{1}{4\pi} \exp\left( -\frac{t^2}{2} - \frac{p^2}{8} \right)$$

---

### 7. Probabilities
* **$P(X \le 0)$:** Since $X \sim N(0,1)$, by symmetry, $P(X \le 0) = \mathbf{0.5}$.
* **$P(X > 1)$:** Using the standard normal table, $1 - \Phi(1) \approx 1 - 0.8413 = \mathbf{0.1587}$.
* **$P(Y \in (-\infty, 0] \times \mathbb{R})$:** This is identical to $P(T \le 0) = \mathbf{0.5}$.
* **$P(Y \in [-1, 1] \times [-2, 2])$:** Due to independence:
  $$P(-1 \le T \le 1) \cdot P(-2 \le P \le 2) = (\Phi(1) - \Phi(-1)) \cdot (\Phi(2/2) - \Phi(-2/2))$$
  $$= (0.6826) \cdot (0.6826) \approx \mathbf{0.466}$$

---

### 8. Conditional Distribution of $X$ given $Y=(t,p)$
If we know $Y$, we know both $t$ and $p$ exactly. Thus, $X$ is no longer random:
$$X | Y=(t,p) \text{ is a point mass (Dirac delta) at } t.$$

### 9. Conditional Distribution of $Y$ given $X=t$
If we know $X=t$, the temperature is fixed, but the pressure $P$ remains random and independent:
$$Y | X=t \sim (t, N(0, 4))$$

### 10. Qualitative Interpretation
* **Conditioning on $X$:** We know the exact horizontal position (temperature) but are completely uncertain about the vertical position (pressure) beyond its prior distribution.
* **Conditioning on $Y$:** We know the exact state of the reactor. No additional information is gained by looking at $X$ once $Y$ is known.

### 11. $X$ as a Measurable Function of $Y$
Since $X(t,p) = \pi_1(Y(t,p))$ and the projection map $\pi_1$ is continuous (and thus Borel measurable), $X$ is a measurable function of $Y$. This proves that the information in $X$ is a subset of the information in $Y$.

---

### 12. Python Visualization
```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal, norm

# Create grid
t = np.linspace(-3, 3, 100)
p = np.linspace(-6, 6, 100)
T, P = np.meshgrid(t, p)
pos = np.dstack((T, P))

# Joint density
rv = multivariate_normal([0, 0], [[1, 0], [0, 4]])
Z = rv.pdf(pos)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Heatmap of Y
ax1.contourf(T, P, Z, cmap='viridis')
ax1.set_title('Joint Density of Y (T and P)')
ax1.set_xlabel('Temperature (T)')
ax1.set_ylabel('Pressure (P)')

# Marginal of X
ax2.plot(t, norm.pdf(t, 0, 1), color='blue', lw=2)
ax2.fill_between(t, norm.pdf(t, 0, 1), alpha=0.3)
ax2.set_title('Marginal Density of X (Temperature)')
ax2.set_xlabel('t')

plt.show()
